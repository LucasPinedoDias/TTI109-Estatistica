```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt

carros = pd.DataFrame({'A': [28, 32, 28, 30, 34],

                       'B': [31, 29, 31, 29, 31],

                       'C': [29, 32, 28, 32, 30]})

display(carros)

display(carros.agg(['mean', 'median']))

display(carros.mode())

display(carros.describe())
plt.show()
```
